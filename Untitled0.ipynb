### 1. Setup & Upload Data

```python
# Core libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Optimization
from scipy.optimize import minimize
```

### 2. Load & Clean Data

```python
# Load data
df = pd.read_csv("nifty50_closing_prices.csv")

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Sort by date
df = df.sort_index()

# Fill missing values
df = df.fillna(method='ffill')

df.head()
```

### 3. Calculate Returns

```python
# Daily returns
daily_returns = df.pct_change().dropna()

daily_returns.head()
```

### 4. Basic Fund (Stock) Performance Metrics

```python
annual_returns = daily_returns.mean() * 252
volatility = daily_returns.std() * np.sqrt(252)

risk_free_rate = 0.06

sharpe_ratio = (annual_returns - risk_free_rate) / volatility

metrics = pd.DataFrame({
    "Annual Return": annual_returns,
    "Volatility": volatility,
    "Sharpe Ratio": sharpe_ratio
})

metrics.sort_values("Sharpe Ratio", ascending=False).head()
```

### 5. Portfolio Construction (User-Defined Weights)

```python
# Equal weight portfolio (simple & explainable)
num_assets = len(daily_returns.columns)
weights = np.ones(num_assets) / num_assets

portfolio_return = np.sum(annual_returns * weights)
portfolio_volatility = np.sqrt(
    np.dot(weights.T, np.dot(daily_returns.cov() * 252, weights))
)

portfolio_sharpe = (portfolio_return - risk_free_rate) / portfolio_volatility

portfolio_return, portfolio_volatility, portfolio_sharpe
```

### 6. Monte Carlo Simulation (Very Simplified)

```python
num_simulations = 2000

returns = []
volatility = []

for i in range(num_simulations):
    weights = np.random.random(num_assets)
    weights /= weights.sum()

    ret = np.sum(annual_returns * weights)
    vol = np.sqrt(np.dot(weights.T, np.dot(daily_returns.cov() * 252, weights)))

    returns.append(ret)
    volatility.append(vol)
```

### 7. Efficient Frontier Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(volatility, returns, alpha=0.4)
plt.xlabel("Risk (Volatility)")
plt.ylabel("Return")
plt.title("Efficient Frontier – NIFTY 50 Portfolio")
plt.show()
```

### 8. Best Portfolio (Max Sharpe Ratio)

```python
def portfolio_stats(weights):
    ret = np.sum(annual_returns * weights)
    vol = np.sqrt(np.dot(weights.T, np.dot(daily_returns.cov() * 252, weights)))
    sharpe = (ret - risk_free_rate) / vol
    return -sharpe  # negative for minimization

constraints = {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
bounds = [(0,1)] * num_assets

optimal = minimize(
    portfolio_stats,
    num_assets * [1/num_assets],
    bounds=bounds,
    constraints=constraints
)

optimal_weights = optimal.x
print(optimal_weights)
```

### 9. Growth of Investment (₹1 Example)

```python
portfolio_daily_returns = daily_returns.dot(optimal_weights)
portfolio_value = (1 + portfolio_daily_returns).cumprod()

plt.figure(figsize=(8,5))
plt.plot(portfolio_value)
plt.title("Growth of ₹1 Investment")
plt.xlabel("Date")
plt.ylabel("Portfolio Value")
plt.show()
```

### 10. Portfolio Composition (Pie Chart)

```python
top_holdings = pd.Series(optimal_weights, index=daily_returns.columns)\
                .sort_values(ascending=False)[:10]

plt.figure(figsize=(7,7))
top_holdings.plot.pie(autopct='%1.1f%%')
plt.title("Top 10 Portfolio Holdings")
plt.ylabel("")
plt.show()
```

### 11. Correlation Heatmap (Diversification Proof)

```python
plt.figure(figsize=(10,8))
sns.heatmap(daily_returns.corr(), cmap="coolwarm")
plt.title("Stock Correlation Heatmap")
plt.show()
```
